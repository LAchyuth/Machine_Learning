```python
from sklearn.preprocessing import LabelEncoder
le =LabelEncoder()

data['Attrition'] = le.fit_transform(data['Attrition'])
data['BusinessTravel'] = le.fit_transform(data['BusinessTravel'])
data['Department'] = le.fit_transform(data['Department'])
data['EducationField'] = le.fit_transform(data['EducationField'])
data['Gender'] = le.fit_transform(data['Gender'])
data['JobRole'] = le.fit_transform(data['JobRole'])
data['Over18'] = le.fit_transform(data['Over18'])
data['OverTime'] = le.fit_transform(data['OverTime'])
data['MaritalStatus'] = le.fit_transform(data['MaritalStatus'])
```

```python
data.head()
```

```python
x = data.drop(['Attrition'],axis =1)
y =data['Attrition']
```

```python
x.head()
```

```python
y.head()
```

```python
y.value_counts()
```

```python
import imblearn
from imblearn.over_sampling import SMOTE
smote = SMOTE()
x_smote, y_smote = smote.fit_resample(x,y)
print(y.value_counts())
print()
print(y_smote.value_counts())
```

```python
plt.figure(figsize = (20,16))
sns.heatmap(data.corr(),annot=True,cmap = 'coolwarm')
plt.show()
```

```python
# split the data into training and test
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test =train_test_split(x_smote,y_smote,test_size =0.25,random_state =101)
```

## RandomForestClassifier

```python
from sklearn.ensemble import RandomForestClassifier
rf = RandomForestClassifier()
rf.fit(x_train,y_train)
```

```python
y_pred_train = rf.predict(x_train)
y_pred_test = rf.predict(x_test)
```

```python
from sklearn.metrics import classification_report,accuracy_score,confusion_matrix
```

```python
print(classification_report(y_train,y_pred_train))
print("**********************"*5)
print(classification_report(y_test,y_pred_test))
```

```python
print(accuracy_score(y_train,y_pred_train))
print("**********************"*5)
print(accuracy_score(y_test,y_pred_test))
```

```python
print(confusion_matrix(y_train,y_pred_train))
print("**********************"*5)
print(confusion_matrix(y_test,y_pred_test))
```

## Cross-Validation

```python
from sklearn.model_selection import cross_val_score
```

```python
training_accuracy = cross_val_score(rf,x_train,y_train,cv =10)
print("Average Accuracy:",training_accuracy.mean())
print("Max Accuracy:",training_accuracy.max())
print("Min Accuracy:",training_accuracy.min())
```

### Model Building again with One Hot Encoding method to compare the performance

```python
data1 = pd.read_csv("D:\Learnbay\ML Real Time Project\HR Analytics\Attrition.csv")
data1.head()
```

```python
data1.info()
```

```python
data1.shape
```

```python
data2 = pd.get_dummies(data1, columns = ['BusinessTravel','Department','EducationField',
                                       'Gender','JobRole','MaritalStatus','Over18','OverTime' ])
```

```python
data2['Attrition'] = data2['Attrition'].astype('category')
data2['Attrition'] = data2['Attrition'].cat.codes
```

```python
data2.shape
```

```python
x = data2.drop(['Attrition'],axis =1)
y = data2['Attrition']
```

```python
import imblearn
from imblearn.over_sampling import SMOTE
smote = SMOTE()

x_smote,y_smote = smote.fit_resample(x,y)
```

```python
print(y.value_counts())
print()
print(y_smote.value_counts())
```

```python
# split the data into training and test
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test =train_test_split(x_smote,y_smote,test_size =0.25,random_state =101)
```

```python
from sklearn.ensemble import RandomForestClassifier
rf = RandomForestClassifier()
rf.fit(x_train,y_train)
```

```python
y_pred_rf_xtrain = rf.predict(x_train)
y_pred_rf_xtest = rf.predict(x_test)
```

```python
print(classification_report(y_train,y_pred_rf_xtrain))
print()
print(classification_report(y_test,y_pred_rf_xtest))
```

```python
print("Training Accuracy:",accuracy_score(y_train,y_pred_rf_xtrain))
print()
print("Test Accuracy:",accuracy_score(y_test,y_pred_rf_xtest))
```

```python
print(confusion_matrix(y_train,y_pred_rf_xtrain))
print()
print(confusion_matrix(y_test,y_pred_rf_xtest))
```

```python
Training_Accuracy = cross_val_score(rf,x_train,y_train)
print(Training_Accuracy.mean())
print(Training_Accuracy.max())
print(Training_Accuracy.min())
```
